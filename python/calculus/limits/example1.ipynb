$$
\begin{equation}
    \lim_{x \rightarrow 1}=\frac{\frac{1}{x}-1}{x-1}
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

x_values = np.arange(-10, 11)
print("x:", x_values)

y_values = list(map(lambda x_value: sp.limit((1 / x_value - 1) / (x_value - 1), x, 1), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(x_values, y_values, label="limit x→1", marker="o", markersize=5)
axes.legend()

plt.show()
```
