[2] Example 2.8
$$
f(x) =
\begin{cases}
\begin{array}
    & x+1 & if x \lt 2 \\
    x^2-4 & if x \geq 2
\end{array}
\end{cases}
$$
$$
\begin{equation}
    \lim_{x \rightarrow 2}=f(x)
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

def f(value):
    if value < 2:
        value += 1
    elif value >= 2:
        value = value ** 2 - 4
    
    return value

x_values = np.array([1.9, 1.99, 1.999, 1.9999, 1.99999, 2.00001, 2.0001, 2.001, 2.01, 2.1])
print("x:", x_values)

y_values = list(map(lambda x_value: sp.limit(f(x_value), x, 2), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(x_values, y_values, label="limit xâ†’2", marker="o", markersize=5)
axes.legend()

plt.show()
```
