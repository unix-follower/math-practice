$$
\begin{equation}
    f(x)=x^3-x^2-3x+1
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = sp.Lambda(x, x ** 3 - x ** 2 - 3 * x + 1)
print(f)

x_values = np.linspace(0, 1, 20)
print("x:", x_values)

y_values = list(map(f, x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(x_values, y_values, label=f.expr, marker="o", markersize=5)
axes.legend()

plt.show()
```
