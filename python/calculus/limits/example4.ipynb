$$
\begin{equation}
    \lim_{x \rightarrow 6}=(2x-1)\sqrt{x+4}
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

left_x_values = np.linspace(5.01, 5.9999, 15)
right_x_values = np.linspace(6.01, 6.9999, 15)
x_values = np.concatenate([left_x_values, right_x_values])
print("x:", x_values)

y_values = list(map(lambda x_value: sp.limit((2 * x_value - 1) * sp.sqrt(x_value + 4), x, 6), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(x_values, y_values, label="limit x→6", marker="o", markersize=5)
axes.legend()

plt.show()
```
