$$
\begin{equation}
    \lim_{\theta \rightarrow 0}=\frac{1-cos \theta}{sin \theta}
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import theta
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

left_theta_values = np.linspace(-0.9999, -0.01, 10)
right_theta_values = np.linspace(0.01, 0.9999, 10)
theta_values = np.concatenate([left_theta_values, right_theta_values])
print("theta:", theta_values)

y_values = list(map(lambda x_value: sp.limit((1 - sp.cos(x_value)) / sp.sin(x_value), theta, 0), theta_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("theta")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(theta_values, y_values, label="limit theta→0", marker="o", markersize=5)
axes.legend()

plt.show()
```
