$$
\begin{equation}
    \lim_{x \rightarrow \pi} sin(x-\pi)
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

left_x_values = np.linspace(2.01, 3.1399, 20)
right_x_values = np.linspace(3.1501, 3.9999, 20)
x_values = np.concatenate([left_x_values, right_x_values])
print("x:", x_values)

y_values = list(map(lambda x_value: sp.limit(sp.sin(x_value - sp.pi), x, sp.pi), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Limit of a function")
axes.grid()
axes.plot(x_values, y_values, label="limit x→π", marker="o", markersize=5)
axes.legend()

plt.show()
```
