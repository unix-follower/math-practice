$$
\begin{gather}
    f(x)=2x-3 \\
    g(x)=x^2-1
\end{gather}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, 2 * x - 3)
print(f)
g = Lambda(x, x**2 - 1)
print(g)

x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(lambda x: f(x) + g(x), x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("(f+g)(x)")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```

```python
print("x:", x_values)
y_values = list(map(lambda x: f(x) - g(x), x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("(f-g)(x)")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```

```python
print("x:", x_values)
y_values = list(map(lambda x: f(x) * g(x), x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("(f*g)(x)")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```

```python
import math

print("x:", x_values)

def map_y_value(x: float):
    y = f(x) / g(x)
    if y.has(sp.zoo):
        return np.inf
    return y
    
y_values = list(map(map_y_value, x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("(f/g)(x)")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```
