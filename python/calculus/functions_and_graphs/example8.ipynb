$$
\begin{equation}
    s(t)=-16t^2+100
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import t
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(t, -16 * t**2 + 100)
print(f)

x_values = np.arange(0, 3, 0.5)
print("x:", x_values)
y_values = list(map(f, x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("time")
axes.set_ylabel("height")
axes.set_title(f.expr)
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```
