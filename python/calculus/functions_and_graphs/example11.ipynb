$$
\begin{gather}
    (g \circ f)(x) \\
    f(x)=x^2+1 \\
    g(x)=1/x
\end{gather}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, x**2 + 1)
print(f)
g = Lambda(x, 1 / x)
print(g)

gof_label = "(g o f)(x)"
print(gof_label)
x_values = np.arange(-5, 10)
print("x:", x_values)
y_values = list(map(lambda x: g(f(x)), x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

figure, axes = plt.subplots()
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.grid()
axes.plot(x_values, y_values, label=gof_label, marker="o", markersize=5)

def map_y_value(y: float):
    if y.has(sp.zoo):
        return np.inf
    return y

x_values = np.arange(-5, 10)
fog_label="(f o g)(x)"
print(fog_label)
print("x:", x_values)
y_values = list(map(lambda x: map_y_value(f(g(x))), x_values))
print("y:", list(map(lambda y: float("{:.2f}".format(y)), y_values)))

axes1 = plt.subplot()
axes1.plot(x_values, y_values, label=fog_label, marker="o", markersize=5)
axes.legend()

plt.show()
```
