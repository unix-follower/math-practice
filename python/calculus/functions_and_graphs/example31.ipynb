$$
\begin{gather}
    f(x)=2x+3 \\
    f^{-1}(y)=\frac{y-3}{2}
\end{gather}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, y
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, 2 * x + 3)
print(f)
f_inverse = Lambda(y, (y - 3) / 2)
print(f_inverse)

x_values = np.arange(-5, 5)
print("x:", x_values)

y_values = list(map(f, x_values))
print("y:", y_values)

y_values_inverse = list(map(f_inverse, y_values))
print("y (inverse):", y_values_inverse)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Inverse Function")
axes.grid()
axes.plot(x_values, y_values, label=f.expr, marker="o", markersize=5)
axes.plot(x_values, y_values_inverse, label=f_inverse.expr, marker="o", markersize=5)
axes.legend()

plt.show()
```
