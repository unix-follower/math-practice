$$
\begin{equation}
    f(x)=\frac{6x^{-3}y^2}{12x^{-4}y^5}
\end{equation}
$$
$$
\begin{equation}
    f(x)=\frac{6x^{-3}y^2}{12x^{-4}y^5} = \frac{6}{12} \cdot x^{(-3+4)} \cdot y^{(2-5)} = \frac{1}{2} \cdot x \cdot y^{-3} = \frac{x}{2y^3}
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, y
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda((x, y), (6 * x**(-3) * y**2) / (12 * x**(-4) * y**5))
print(f)

x_values = np.arange(0, 11)
print("x:", x_values)

y_values = list(map(lambda x: f(x, x + 1), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Exponential function")
axes.grid()
axes.plot(x_values, y_values, label=f.expr, marker="o", markersize=5)
axes.legend()

plt.show()
```
