$$
f(x) =
\begin{cases}
\begin{array}
    & 2-x, & x \leq 2 \\
    x+2, & x \gt 2
\end{array}
\end{cases}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

leq2_f = Lambda(x, 2-x)
print(leq2_f)

x_values = np.arange(-5, 3)
print("x:", x_values)

y_values = list(map(leq2_f, x_values))
print("y:", y_values)

axes = plt.subplot(211)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Piecewise-Defined Function")
axes.grid()
leq2_line, = axes.plot(x_values, y_values, marker="o", markersize=5)

gt2_f = Lambda(x, x+2)
print(gt2_f)

x_values = np.arange(3, 11)
print("x:", x_values)

y_values = list(map(gt2_f, x_values))
print("y:", y_values)

axes = plt.subplot(211)
gt2_line, = axes.plot(x_values, y_values, marker="o", markersize=5)

axes.legend(labels=[leq2_f.expr, gt2_f.expr], handles=[leq2_line, gt2_line])
plt.show()
```
