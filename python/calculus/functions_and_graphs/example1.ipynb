```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, a
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, 3 * x**2 + 2 * x - 1)
print(f)

result = f(-2)
assert 7 == result, f"The actual value: {result}"

result = f(sp.sqrt(2))
assert "2*sqrt(2) + 5" == str(result), f"The actual value: {result}"

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title(result)
axes.plot(result, marker="o", markersize=5)
plt.show()

h = sp.symbols("h")
result = f(a + h)
assert "2*a + 2*h + 3*(a + h)**2 - 1" == str(result), f"The actual value: {result}"

x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(lambda value: f(value + 2 * value), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title(result)
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```
