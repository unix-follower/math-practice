$$
\begin{gather}
    f(x)=2-5x \\
    g(x)=\sqrt x \\
    (f \circ g)(x) \\
\end{gather}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, 2 - 5 * x)
print(f)
g = Lambda(x, sp.sqrt(x))
print(g)

x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(lambda x: f(g(x)), x_values))
print("y:", y_values)

figure, axes = plt.subplots()
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("(f o g)(x)")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)

plt.show()
```
