$$
\begin{equation}
    f(x)=100*3^{x/2}
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, 100 * 3**(x / 2))
print(f)

x_values = np.arange(0, 11)
print("x:", x_values)

y_values = list(map(f, x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Exponential function")
axes.grid()
axes.plot(x_values, y_values, label=f.expr, marker="o", markersize=5)
axes.legend()

plt.show()
```
