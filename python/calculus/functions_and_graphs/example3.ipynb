$$
\begin{equation}
    ùëì(ùë•)=(ùë•-4)^2+5
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, a, h
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, (x - 4)**2 + 5)
print(f)

x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(f, x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_xticks(np.arange(0, 10))
axes.set_yticks(np.arange(0, 40, 5))
axes.set_title(f.expr)
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```

$$
\begin{equation}
    ùëì(ùë•)=\sqrt{3ùë•}+2-1
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, a, h
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

f = Lambda(x, sp.sqrt(3 * x) + 2 - 1)
print(f)

x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(f, x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_xticks(np.arange(0, 20))
axes.set_yticks(np.arange(0, 10))
axes.set_title(f.expr)
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```

$$
\begin{equation}
    ùëì(ùë•)=\frac{3}{ùë•-2}
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, a, h
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

f = Lambda(x, 3 / (x - 2))
print(f)


def map_x_value(x_value: int):
    if x_value == 2:
        return np.inf
    else:
        return f(x_value)


x_values = np.arange(0, 10)
print("x:", x_values)
y_values = list(map(map_x_value, x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_xticks(np.arange(-2, 10))
axes.set_yticks(np.arange(-5, 5))
axes.set_title(f.expr)
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```
