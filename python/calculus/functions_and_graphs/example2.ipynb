```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x, a, h
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, x**2 - 3 * x + 5)
print(f)

result = f(1)
assert 3 == result, f"The actual value: {result}"

figure, axes = plt.subplots(figsize=(6, 2))
axes: plt.Axes
axes.plot(result, marker="o", markersize=5)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("f(1)=3")
axes.set_yticks(np.arange(0, 4.5))
plt.show()

h = sp.symbols("h")
result = f(a + h)
assert "-3*a - 3*h + (a + h)**2 + 5" == str(result), f"The actual value: {result}"

x_values = np.arange(0, 5)
print("x:", x_values)
y_values = list(map(lambda value: result.subs({"a": value, "h": 2 * value}), x_values))
print("y:", y_values)

figure, axes = plt.subplots(figsize=(6, 2))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title(result)
axes.plot(x_values, y_values, marker="o", markersize=5)
plt.show()
```
