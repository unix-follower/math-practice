$$
\begin{equation}
    1+cos(2\theta)=cos(\theta)
\end{equation}
$$

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

import sys
sys.path.insert(1, "../../utils")

from basic_units import degrees, radians

print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()


def create_axes_values(is_arange_theta_values=False):
    if is_arange_theta_values:
        theta = np.arange(0, 2, 0.09)
    else:
        n = 0 # 0, +/-1, +/-2
        theta = np.array([
            np.pi / 2 + n * np.pi,
            np.pi / 3 + 2 * n * np.pi,
            -np.pi / 3 + 2 * n * np.pi,
        ])
        
    print(theta)
    radial_distance = 1 + np.cos(2 * theta)
    print(radial_distance)
    
    rhs_radial_distance = np.cos(theta)
    print(rhs_radial_distance)
    are_equal = np.isclose(rhs_radial_distance, radial_distance)
    print(are_equal)

    return theta, radial_distance


theta, radial_distance = create_axes_values()
x = [val*radians for val in theta]

figure, (axes1, axes2) = plt.subplots(2)
axes1.plot(x, radial_distance, xunits=radians)
axes2.plot(x, radial_distance, xunits=degrees)

figure = plt.figure()
axes = figure.add_subplot(projection="polar")
axes.plot(theta, radial_distance)
axes.grid(True)

plt.show()
```

$$
\begin{equation}
    sin(2\theta)=tan(\theta)
\end{equation}
$$

```python
def create_axes_values(is_arange_theta_values=False):
    if is_arange_theta_values:
        theta = np.arange(0, 2, 0.09)
    else:
        n = 1 # 0, +/-1, +/-2
        theta = np.array([
            n * np.pi,
            np.pi / 4 + (n * np.pi) / 2 
        ])

    print(theta)
    radial_distance = np.sin(2*theta)
    print(radial_distance)
    
    rhs_radial_distance = np.tan(theta)
    print(rhs_radial_distance)
    are_equal = np.isclose(rhs_radial_distance, radial_distance)
    print(are_equal)

    return theta, radial_distance

theta, radial_distance = create_axes_values()
x = [val*radians for val in theta]

figure, (axes1, axes2) = plt.subplots(2)
axes1.plot(x, radial_distance, xunits=radians)
axes2.plot(x, radial_distance, xunits=degrees)

figure = plt.figure()
axes = figure.add_subplot(projection="polar")
axes.plot(theta, radial_distance)
axes.grid(True)

plt.show()
```
