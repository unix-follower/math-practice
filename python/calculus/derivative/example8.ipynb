$$
f(x)=
\begin{cases}
\begin{array}
& ax+b & if \space x \lt 3 \\
x^2 & if \space x \ge 3
\end{array}
\end{cases}
$$

```python
import sympy as sp
from sympy.abc import a, b, x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = sp.Piecewise((a * x + b, x < 3), (x ** 2, x >= 3))
print(f)

a_coefficient = 6
b_coefficient = -9
x_values = np.arange(-10, 11)
print(f"x:{x_values}")

def calculate_y_values(x_value: float):
    return f.subs(x, x_value).subs(a, a_coefficient).subs(b, b_coefficient)
    
y_values = list(map(calculate_y_values, x_values))
print(f"y:{y_values}")

f_prime = f.diff(x)
print(f"f′:{f_prime}")
x_at = -3
slope = f_prime.subs(x, x_at).subs(a, a_coefficient).subs(b, b_coefficient).evalf()
print(f"The slope of the tangent at x={x_at}: {slope}")
y_tangent = slope * (x_values - x_at) + f.subs(x, x_at).subs(a, a_coefficient).subs(b, b_coefficient).evalf()
print(f"y_tangent:{y_tangent}")

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Differentiable piecewise function")
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
axes.plot(x_values, y_tangent, label=f"The slope at x={x_at} (slope={slope:.2f})", linestyle="--")
axes.legend()
plt.show()
```
