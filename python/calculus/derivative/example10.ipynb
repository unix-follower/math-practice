$$
\begin{equation}
s(t)=t^3
\end{equation}
$$

```python
import sympy as sp
from sympy.abc import t
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = sp.Lambda(t, t ** 3)
print(f)

t_values = np.arange(0, 7)
print(f"t:{t_values}")
y_values = list(map(f, t_values))
print(f"y:{y_values}")

f_prime = sp.diff(f(t), t)
print(f"f′:{f_prime}")

f2_derivative = sp.diff(f_prime, t)
print(f"f′′:{f2_derivative}")

t_at = 3

slope = f_prime.subs(t, t_at).evalf()
print(f"f′ at t={t_at}: {slope}")
y_tangent = slope * (t_values - t_at) + f(t_at).evalf()
print(f"y_tangent:{y_tangent}")

slope2 = f2_derivative.subs(t, t_at).evalf()
print(f"f′′ at t={t_at}: {slope2}")
y2_tangent = slope2 * (t_values - t_at) + f(t_at).evalf()
print(f"y2_tangent:{y2_tangent}")

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("time (seconds)")
axes.set_ylabel("y")
axes.set_title(f.expr)
axes.grid()
axes.plot(t_values, y_values, marker="o", markersize=5)
axes.plot(t_values, y_tangent, label=f"f′ at t={t_at} (slope={slope:.2f})", linestyle="--")
axes.plot(t_values, y2_tangent, label=f"f′′ at t={t_at} (slope={slope:.2f})", linestyle="-.")
axes.legend()
plt.show()
```
