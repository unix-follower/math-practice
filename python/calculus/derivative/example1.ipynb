$$
\begin{equation}
f(x)=\sqrt{x}
\end{equation}
$$
___
$$
\begin{equation}
\lim_{h \rightarrow 0} \frac{f(4+h)-f(4)}{h} = \frac{1}{4}
\end{equation}
$$
$$
\begin{equation}
\lim_{h \rightarrow 0} \frac{\sqrt{4+h} - \sqrt{4}}{h}
\end{equation}
$$
$$
\begin{equation}
\lim_{h \rightarrow 0} \frac{\sqrt{4+h} - \sqrt{4}}{h} \cdot \frac{\sqrt{4+h} + \sqrt{4}}{\sqrt{4+h} + \sqrt{4}}
\end{equation}
$$
$$
\begin{equation}
(\sqrt{4+h} - \sqrt{4}) \cdot (\sqrt{4+h} + \sqrt{4}) = (\sqrt{4+h})^2 - (\sqrt{4})^2
\end{equation}
$$
$$
\begin{equation}
(\sqrt{4+h})^2 - (\sqrt{4})^2 = (4 + h) - 4 = h
\end{equation}
$$
$$
\begin{equation}
\lim_{h \rightarrow 0} \frac{h}{h(\sqrt{4+h} + \sqrt{4})}
\end{equation}
$$
$$
\begin{equation}
\lim_{h \rightarrow 0} \frac{1}{\sqrt{4+h} + \sqrt{4}}
\end{equation}
$$
$$
\begin{equation}
\frac{1}{\sqrt{4+0} + \sqrt{4}} = \frac{1}{\sqrt{4} + \sqrt{4}} = \frac{1}{2\sqrt{4}} = \frac{1}{4}
\end{equation}
$$

```python
import sympy as sp
from sympy import Lambda
from sympy.abc import x
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print()

f = Lambda(x, sp.sqrt(x))
print(f)

x_values = np.arange(0, 5)
print(f"x:{x_values}")
y_values = list(map(f, x_values))
print(f"y:{y_values}")

f_prime = sp.diff(f(x), x)
x_at = 4
slope = f_prime.subs(x, x_at).evalf()
print(f"The slope of the tangent at x={x_at}: {slope}")
y_tangent = slope * (x_values - x_at) + f(x_at).evalf()
print(f"y_tangent:{y_tangent}")

figure, axes = plt.subplots(figsize=(6, 3))
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title(f.expr)
axes.grid()
axes.plot(x_values, y_values, marker="o", markersize=5)
axes.plot(x_values, y_tangent, label=f"The slope at x={x_at} (slope={slope:.2f})", linestyle="--")
axes.legend()
plt.show()
```
