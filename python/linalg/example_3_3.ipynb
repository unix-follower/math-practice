$$

3x+4y=3 \\
2x-y=1
\end{gather}
$$

```python
import sympy as sp

print(f"SymPy version: {sp.__version__}")

x, y = sp.symbols("x, y")

equation1 = 3 * x + 4 * y - 3
equation2 = 2 * x - y - 1

result: dict = sp.solve([equation1, equation2], (x, y))
print(type(result))
print(result)

assert "7/11" == str(result[x]), f"The actual value: {result[x]}"
assert "3/11" == str(result[y]), f"The actual value: {result[y]}"
```
