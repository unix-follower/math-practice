$$
\begin{gather}
x-y=0 \\
2x-2y+z+2w=4 \\
y+w=0 \\
2z+w=5
\end{gather}
$$

```python
import sympy as sp

print(f"SymPy version: {sp.__version__}")

x, y, z, w = sp.symbols("x, y, z, w")

equation1 = x - y
equation2 = 2 * x - 2 * y + z + 2 * w - 4
equation3 = y + w
equation4 = 2 * z + w - 5

result: dict = sp.solve([equation1, equation2, equation3, equation4], (x, y, z, w))
print(type(result))
print(result)

assert -1 == result[x], f"The actual value: {result[x]}"
assert -1 == result[y], f"The actual value: {result[y]}"
assert 2 == result[z], f"The actual value: {result[z]}"
assert 1 == result[w], f"The actual value: {result[w]}"
```
