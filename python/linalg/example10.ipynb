$$

x+y=4 \\
2x+2y=8
\end{gather}
$$

```python
import sympy as sp

print(f"SymPy version: {sp.__version__}")

x, y = sp.symbols("x, y")

equation1 = x + y - 4
equation2 = 2 * x + 2 * y - 8

result: dict = sp.solve([equation1, equation2], (x, y))
print(type(result))
print(result)

assert 4 - y == result[x], "There are many solutions"
```
