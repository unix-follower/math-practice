$$
\begin{gather}
x+y+z=9 \\
2x+4y-3z=1 \\
3x+6y-5z=0
\end{gather}
$$

```python
import sympy as sp

print(f"SymPy version: {sp.__version__}")

x, y, z = sp.symbols("x, y, z")

equation1 = x + y + z - 9
equation2 = 2 * x + 4 * y - 3 * z - 1
equation3 = 3 * x + 6 * y - 5 * z

result: dict = sp.solve([equation1, equation2, equation3], (x, y, z))
print(type(result))
print(result)

assert 7 == result[x], f"The actual value: {result[x]}"
assert -1 == result[y], f"The actual value: {result[y]}"
assert 3 == result[z], f"The actual value: {result[z]}"
```
